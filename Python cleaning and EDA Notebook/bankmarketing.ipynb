**Loading and Checking data**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/bank.csv")

# Preview first rows
print(df.head())

# Number of rows and columns
print("\nShape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types + missing values
print("\nInfo:")
print(df.info())

# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

**Missing and Unknown Values**

```python
#handling missing data
print (df.isnull().sum())
```

```python
print((df == "unknown").sum())
```

```python
#convert unknown into missing
import numpy as np
print (df.replace("unknown", np.nan, inplace=True))
```

```python
print (df.isnull().sum())
```

```python
# fixing missing values
df["job"] = df["job"].fillna("Not Specified")
df["education"] = df["education"].fillna("Not Specified")
df["contact"] = df["contact"].fillna("Not Specified")
df["poutcome"] = df["poutcome"].fillna("No Previous Campaign")
```

```python
print (df.isnull().sum())
```

**Duplicates**

```python
#checking for duplicates
print (df.duplicated().sum())
```

**Checking the Pdays**

```python
# pdays count
df["pdays"].value_counts()
```

```python
#convert -1 as missing
import numpy as np
df["pdays"] = df["pdays"].replace(-1, np.nan)
```

```python
df["pdays"].isnull().sum()
```

```python
# creating contact history
df["previously_contacted"] = df["pdays"].notnull()
```

```python
df["previously_contacted"].value_counts()
```
